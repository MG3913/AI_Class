```python
# 결측치 확인
df.isnull().sum()
```

```python
#위에서부터 임신, 포도당, 혈압, 피부두께, 인슐린, BMI지수, 
#당뇨병 혈통함수(당뇨병을 앓은 혈연), 나이, 결과
#결과는 Y(실제 값)
#나머지 결측치는 모두 제거하면 안됨. 따라서 평균값으로 대체. 
mean_Pregnancies = df['Pregnancies'].mean()
mean_Glucose = df['Glucose'].mean()
mean_BloodPressure = df['BloodPressure'].mean()
mean_SkinThickness  = df['SkinThickness'].mean()
mean_Insulin = df['Insulin'].mean()
mean_BMI = df['BMI'].mean()
mean_DiabetesPedigreeFunction = df['DiabetesPedigreeFunction'].mean()
mean_Age = df['Age'].mean()
```

```python
df['Pregnancies'].fillna(mean_Pregnancies, inplace=True)
df['Glucose'].fillna(mean_Glucose, inplace=True)
df['BloodPressure'].fillna(mean_BloodPressure, inplace=True)
df['SkinThickness'].fillna(mean_SkinThickness, inplace=True)
df['Insulin'].fillna(mean_Insulin, inplace=True)
df['BMI'].fillna(mean_BMI, inplace=True)
df['DiabetesPedigreeFunction'].fillna(mean_DiabetesPedigreeFunction, inplace=True)
df['Age'].fillna(mean_Age, inplace=True)
df
```

```python
df.isnull().sum()
```

```python
#인슐린, BMI, 포도당, 혈압, 피부두께(굳기?)는 0이 될 수 없음. 
#따라서 0인 행을 outlier로 인식하고 제거

columns_with_outliers = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']
df = df[(df[columns_with_outliers] != 0).all(axis=1)]
df
```

```python
#인덱스 리셋
from sklearn.preprocessing import StandardScaler
df.reset_index(drop=True, inplace=True)
```

```python
# 스케일러 초기화
scaler = StandardScaler()
# 'Outcome' 열을 제외한 데이터 스케일링
scaled_features = scaler.fit_transform(df.drop('Outcome', axis=1))
```

```python
# 스케일된 데이터를 DataFrame으로 변환
scaled_df = pd.DataFrame(scaled_features, columns=df.columns[:-1])

# 'Outcome' 열 추가
scaled_df['Outcome'] = df['Outcome'].values
```
